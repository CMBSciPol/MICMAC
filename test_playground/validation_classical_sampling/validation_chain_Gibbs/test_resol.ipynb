```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import healpy as hp
import astropy.io.fits as fits
from matplotlib.colors import LogNorm, SymLogNorm
```

```python
nside = 64
lmax = 2*nside
n_iter = 8


pixsize = hp.nside2resol(nside)
pixsize_perso = 4*np.pi/(12*nside**2)

pix_test = np.random.normal(loc=0, scale=1, size=(3, 12*nside**2))
pix_test_2 = np.random.normal(loc=0, scale=1/pixsize, size=(3, 12*nside**2))

test_anafast = hp.anafast(pix_test, lmax=lmax, iter=n_iter)
# test_anafast_2 = hp.anafast(pix_test, lmax=3*nside-1, iter=n_iter)
test_alm = hp.sphtfunc.synalm(test_anafast, lmax=lmax)
test_alm[:] = np.random.normal(loc=0, scale=1, size=test_alm.shape)

test_anafast_alt = hp.anafast(pix_test_2, lmax=lmax, iter=n_iter)

test_anafast_alms = hp.alm2cl(test_alm, lmax=lmax)


test_alm_2 = hp.map2alm(pix_test, lmax=lmax, iter=n_iter)
test_cls_alm_2 = hp.alm2cl(test_alm_2, lmax=lmax)
```

```python

ell_arange = np.arange(test_anafast_alt.shape[1])
plt.plot(ell_arange, test_anafast_alt.T)

plt.plot([ell_arange[0],ell_arange[-1]], [pixsize**2, pixsize**2])
plt.plot([ell_arange[0],ell_arange[-1]], [pixsize_perso, pixsize_perso], '--')

plt.loglog()
plt.show()
```

```python

ell_arange = np.arange(test_anafast.shape[1])
plt.plot(ell_arange, test_anafast.T)

pixsize = hp.nside2resol(nside)
pixsize_perso = 4*np.pi/(12*nside**2)
plt.plot([ell_arange[0],ell_arange[-1]], [pixsize**2, pixsize**2])
plt.plot([ell_arange[0],ell_arange[-1]], [pixsize_perso, pixsize_perso], '--')

plt.loglog()
plt.show()
```

```python

ell_arange = np.arange(test_anafast.shape[1])
plt.plot(ell_arange, test_anafast.T)

pixsize = hp.nside2resol(nside)
plt.plot([ell_arange[0],ell_arange[-1]], [pixsize**2, pixsize**2])
plt.loglog()
plt.show()
```
