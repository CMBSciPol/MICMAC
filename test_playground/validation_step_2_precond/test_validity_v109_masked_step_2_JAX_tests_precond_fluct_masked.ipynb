```python
@jax.custom_jvp
def f(x, y):
  return jnp.sin(x) * y

# f.nondiff_argnums = (2,)

@f.defjvp
def f_jvp(primals, tangents):
  x, y = primals
  x_dot, y_dot = tangents
  primal_out = f(x, y)
  tangent_out = jnp.cos(x) * x_dot * y + jnp.sin(x) * y_dot
  return primal_out, tangent_out
```
