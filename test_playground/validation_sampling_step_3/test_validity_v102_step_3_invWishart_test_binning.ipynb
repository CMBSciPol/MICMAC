```python
    def get_binned_red_c_ells(red_c_ells_to_bin):
        """ Bin the power spectrum to get the binned power spectrum
            
                Parameters
                ----------
                :param red_c_ells_to_bin: power spectrum to bin ; must be of dimension [lmax+1, nstokes, nstokes]

                Returns
                -------
                :return: Binned power spectrum, of dimension [number_bins, nstokes, nstokes]
        """
        chx.assert_axis_dimension(red_c_ells_to_bin, 0, MICMAC_sampler_obj.lmax+1 - MICMAC_sampler_obj.lmin)

        ell_distribution = jnp.arange(red_c_ells_to_bin.shape[0]) + MICMAC_sampler_obj.lmin

        number_bins = MICMAC_sampler_obj.bin_ell_distribution.shape[0]-1
        nstokes = red_c_ells_to_bin.shape[1]

        def map_binned_red_c_ells(bin_ell):
            """ Compute the binned power spectrum for a given ell
            """
            cond = jnp.logical_and(MICMAC_sampler_obj.bin_ell_distribution[bin_ell]<=ell_distribution, MICMAC_sampler_obj.bin_ell_distribution[bin_ell+1]>ell_distribution)
            cond_quantitative = jnp.where(cond, 1, 0)
            modes_to_bin = jnp.einsum('lij, l->lij', red_c_ells_to_bin, cond_quantitative)
            # return jnp.sum(red_c_ells_to_bin.at[MICMAC_sampler_obj.bin_ell_distributed_min[bin_ell]:MICMAC_sampler_obj.bin_ell_distributed_max[bin_ell]+1].get(), axis=0)/(MICMAC_sampler_obj.bin_ell_distributed_max[bin_ell]-MICMAC_sampler_obj.bin_ell_distributed_min[bin_ell]+1)
            return modes_to_bin.sum(axis=0)#/(MICMAC_sampler_obj.bin_ell_distribution[bin_ell+1]-MICMAC_sampler_obj.bin_ell_distribution[bin_ell])

        binned_red_c_ells = jax.vmap(map_binned_red_c_ells)(jnp.arange(number_bins))
        return binned_red_c_ells


    def get_binned_inverse_wishart_sampling_from_c_ells(sigma_ell, PRNGKey):
        """ Solve sampling step 3 : inverse Wishart distribution with C

            sigma_ell is expected to be exactly the parameter of the inverse Wishart (so it should NOT be multiplied by 2*ell+1 if it is thought as a power spectrum)

            Compute a matrix sample following an inverse Wishart distribution. The 3 steps follow Gupta & Nagar (2000) :
                1. Sample n = 2*ell - p + 2*q_prior independent Gaussian vectors with covariance (sigma_ell)^{-1}
                2. Compute their outer product to form a matrix of dimension n_stokes*n_stokes ; which gives us a sample following the Wishart distribution
                3. Invert this matrix to obtain the final result : a matrix sample following an inverse Wishart distribution

            Also assumes the monopole and dipole to be 0

            Parameters
            ----------
            :param sigma_ell: initial power spectrum which will define the parameter matrix of the inverse Wishart distribution ; must be of dimension [number_correlations, lmax+1]
            :param PRNGKey: random key for JAX PNRG
            
            Returns
            -------
            :return: Matrices following an inverse Wishart distribution, of dimensions [lmin:lmax, nstokes, nstokes]
        """

        # chx.assert_axis_dimension(sigma_ell, 1, MICMAC_sampler_obj.lmax+1)
        chx.assert_axis_dimension(sigma_ell, 1, MICMAC_sampler_obj.lmax+1-MICMAC_sampler_obj.lmin)
        # c_ells_Wishart_modified = jnp.copy(sigma_ell)*(2*jnp.arange(MICMAC_sampler_obj.lmax+1) + 1)
        c_ells_Wishart_modified = jnp.copy(sigma_ell)*(2*(jnp.arange(MICMAC_sampler_obj.lmax+1-MICMAC_sampler_obj.lmin)+MICMAC_sampler_obj.lmin) + 1)
        # invert_parameter_Wishart = jnp.linalg.pinv(get_reduced_matrix_from_c_ell_jax(c_ells_Wishart_modified))
        binned_invert_parameter_Wishart = jnp.linalg.pinv(get_binned_red_c_ells(get_reduced_matrix_from_c_ell_jax(c_ells_Wishart_modified)))
        
        
        # sampling_Wishart = jnp.zeros_like(invert_parameter_Wishart)
        sampling_Wishart = jnp.zeros_like(binned_invert_parameter_Wishart)
        # number_dof = lambda x: 2*x+1

        def map_sampling_Wishart(ell_PNRGKey, binned_ell):
            """ Compute the sampling of the Wishart distribution for a given ell
            """

            # sample_gaussian = random.multivariate_normal(ell_PNRGKey, jnp.zeros(MICMAC_sampler_obj.nstokes), invert_parameter_Wishart[ell], shape=(2*(MICMAC_sampler_obj.lmax+1) - MICMAC_sampler_obj.nstokes,))
            # sample_gaussian = random.multivariate_normal(ell_PNRGKey, jnp.zeros(MICMAC_sampler_obj.nstokes), binned_invert_parameter_Wishart[binned_ell], shape=(number_dof(MICMAC_sampler_obj.lmax) - MICMAC_sampler_obj.nstokes-1,))
            # sample_gaussian = random.multivariate_normal(ell_PNRGKey, jnp.zeros(MICMAC_sampler_obj.nstokes), binned_invert_parameter_Wishart[binned_ell], shape=(max_number_dof - MICMAC_sampler_obj.nstokes-1,))
            sample_gaussian = random.multivariate_normal(ell_PNRGKey, jnp.zeros(MICMAC_sampler_obj.nstokes), binned_invert_parameter_Wishart[binned_ell], shape=(MICMAC_sampler_obj.maximum_number_dof - MICMAC_sampler_obj.nstokes-1,))

            # weighting = jnp.where(ell >= (jnp.arange(2*(MICMAC_sampler_obj.lmax+1)-MICMAC_sampler_obj.nstokes)+MICMAC_sampler_obj.nstokes)/2, 1, 0)
            # weighting = jnp.where(ell >= (jnp.arange(2*(MICMAC_sampler_obj.lmax)-MICMAC_sampler_obj.nstokes)+MICMAC_sampler_obj.nstokes)/2, 1, 0)
            # weighting = jnp.where(number_dof(binned_ell)-MICMAC_sampler_obj.nstokes-1 >= jnp.arange(number_dof(MICMAC_sampler_obj.lmax)-MICMAC_sampler_obj.nstokes-1), 1, 0)
            # weighting = jnp.where(number_dof(binned_ell)-MICMAC_sampler_obj.nstokes-1 >= jnp.arange(max_number_dof-MICMAC_sampler_obj.nstokes-1), 1, 0)
            weighting = jnp.where(number_dof(binned_ell)-MICMAC_sampler_obj.nstokes-1 >= jnp.arange(MICMAC_sampler_obj.maximum_number_dof-MICMAC_sampler_obj.nstokes-1), 1, 0)
            

            sample_to_return = jnp.einsum('lk,l,lm->km',sample_gaussian,weighting,sample_gaussian)
            # new_carry = new_ell_PRNGKey
            return sample_to_return

        # distribution_bins_over_lmin = MICMAC_sampler_obj.bin_ell_distributed_min[MICMAC_sampler_obj.bin_ell_distributed_min>=MICMAC_sampler_obj.lmin]
        distribution_bins_over_lmin = jnp.copy(MICMAC_sampler_obj.bin_ell_distribution)[:-1]

        # number_bins_over_lmin = distribution_bins_over_lmin.shape[0]
        number_bins_over_lmin = jnp.size(distribution_bins_over_lmin)

        # PRNGKey_map = random.split(PRNGKey, MICMAC_sampler_obj.lmax-MICMAC_sampler_obj.lmin+1) # Prepare lmax+1-lmin PRNGKeys to be used
        PRNGKey_map = random.split(PRNGKey, number_bins_over_lmin) # Prepare lmax+1-lmin PRNGKeys to be used

        
        # sampling_Wishart_map = jax.vmap(map_sampling_Wishart)(PRNGKey_map, jnp.arange(MICMAC_sampler_obj.lmin,MICMAC_sampler_obj.lmax+1))
        sampling_Wishart_map = jax.vmap(map_sampling_Wishart)(PRNGKey_map, jnp.arange(number_bins_over_lmin))
        # Map over PRNGKeys and ells to create samples of the Wishart distribution, of dimension [lmax+1-lmin,nstokes,nstokes]

        sampling_Wishart = sampling_Wishart.at[-number_bins_over_lmin:].set(sampling_Wishart_map)
        return jnp.linalg.pinv(sampling_Wishart)

jitted_get_binned_inverse_wishart_sampling_from_c_ells = jax.jit(get_binned_inverse_wishart_sampling_from_c_ells)
```

```python
    def get_binned_red_c_ells_v2(red_c_ells_to_bin):
        """ Bin the power spectrum to get the binned power spectrum
            
                Parameters
                ----------
                :param red_c_ells_to_bin: power spectrum to bin ; must be of dimension [lmax+1, nstokes, nstokes]

                Returns
                -------
                :return: Binned power spectrum, of dimension [number_bins, nstokes, nstokes]
        """
        chx.assert_axis_dimension(red_c_ells_to_bin, 0, MICMAC_sampler_obj.lmax+1 - MICMAC_sampler_obj.lmin)

        ell_distribution = jnp.arange(red_c_ells_to_bin.shape[0]) + MICMAC_sampler_obj.lmin

        number_bins = MICMAC_sampler_obj.bin_ell_distribution.shape[0]-1
        nstokes = red_c_ells_to_bin.shape[1]

        def map_binned_red_c_ells(bin_ell):
            """ Compute the binned power spectrum for a given ell
            """
            cond = jnp.logical_and(MICMAC_sampler_obj.bin_ell_distribution[bin_ell]<=ell_distribution, MICMAC_sampler_obj.bin_ell_distribution[bin_ell+1]>ell_distribution)
            cond_quantitative = jnp.where(cond, 1, 0)
            modes_to_bin = jnp.einsum('lij, l, l->lij', red_c_ells_to_bin, cond_quantitative, ell_distribution*(ell_distribution+1))

            return modes_to_bin.sum(axis=0)#/(MICMAC_sampler_obj.bin_ell_distribution[bin_ell+1]-MICMAC_sampler_obj.bin_ell_distribution[bin_ell])

        binned_red_c_ells = jax.vmap(map_binned_red_c_ells)(jnp.arange(number_bins))
        return binned_red_c_ells


    def get_binned_inverse_wishart_sampling_from_c_ells_v2(sigma_ell, PRNGKey):
        """ Solve sampling step 3 : inverse Wishart distribution with C

            sigma_ell is expected to be exactly the parameter of the inverse Wishart (so it should NOT be multiplied by 2*ell+1 if it is thought as a power spectrum)

            Compute a matrix sample following an inverse Wishart distribution. The 3 steps follow Gupta & Nagar (2000) :
                1. Sample n = 2*ell - p + 2*q_prior independent Gaussian vectors with covariance (sigma_ell)^{-1}
                2. Compute their outer product to form a matrix of dimension n_stokes*n_stokes ; which gives us a sample following the Wishart distribution
                3. Invert this matrix to obtain the final result : a matrix sample following an inverse Wishart distribution

            Also assumes the monopole and dipole to be 0

            Parameters
            ----------
            :param sigma_ell: initial power spectrum which will define the parameter matrix of the inverse Wishart distribution ; must be of dimension [number_correlations, lmax+1]
            :param PRNGKey: random key for JAX PNRG
            
            Returns
            -------
            :return: Matrices following an inverse Wishart distribution, of dimensions [lmin:lmax, nstokes, nstokes]
        """

        # chx.assert_axis_dimension(sigma_ell, 1, MICMAC_sampler_obj.lmax+1)
        chx.assert_axis_dimension(sigma_ell, 1, MICMAC_sampler_obj.lmax+1-MICMAC_sampler_obj.lmin)
        # c_ells_Wishart_modified = jnp.copy(sigma_ell)*(2*jnp.arange(MICMAC_sampler_obj.lmax+1) + 1)
        c_ells_Wishart_modified = jnp.copy(sigma_ell)*(2*(jnp.arange(MICMAC_sampler_obj.lmax+1-MICMAC_sampler_obj.lmin)+MICMAC_sampler_obj.lmin) + 1)
        # invert_parameter_Wishart = jnp.linalg.pinv(get_reduced_matrix_from_c_ell_jax(c_ells_Wishart_modified))
        binned_invert_parameter_Wishart = jnp.linalg.pinv(get_binned_red_c_ells_v2(get_reduced_matrix_from_c_ell_jax(c_ells_Wishart_modified)))
        
        
        # sampling_Wishart = jnp.zeros_like(invert_parameter_Wishart)
        sampling_Wishart = jnp.zeros_like(binned_invert_parameter_Wishart)
        # number_dof = lambda x: 2*x+1

        def map_sampling_Wishart(ell_PNRGKey, binned_ell):
            """ Compute the sampling of the Wishart distribution for a given ell
            """

            sample_gaussian = random.multivariate_normal(ell_PNRGKey, jnp.zeros(MICMAC_sampler_obj.nstokes), binned_invert_parameter_Wishart[binned_ell], shape=(MICMAC_sampler_obj.maximum_number_dof - MICMAC_sampler_obj.nstokes-1,))

            weighting = jnp.where(number_dof(binned_ell)-MICMAC_sampler_obj.nstokes-1 >= jnp.arange(MICMAC_sampler_obj.maximum_number_dof-MICMAC_sampler_obj.nstokes-1), 1, 0)

            sample_to_return = jnp.einsum('lk,l,lm->km',sample_gaussian,weighting,sample_gaussian)
            # new_carry = new_ell_PRNGKey
            return sample_to_return

        # distribution_bins_over_lmin = MICMAC_sampler_obj.bin_ell_distributed_min[MICMAC_sampler_obj.bin_ell_distributed_min>=MICMAC_sampler_obj.lmin]
        distribution_bins_over_lmin = jnp.copy(MICMAC_sampler_obj.bin_ell_distribution)[:-1]

        # number_bins_over_lmin = distribution_bins_over_lmin.shape[0]
        number_bins_over_lmin = jnp.size(distribution_bins_over_lmin)

        # PRNGKey_map = random.split(PRNGKey, MICMAC_sampler_obj.lmax-MICMAC_sampler_obj.lmin+1) # Prepare lmax+1-lmin PRNGKeys to be used
        PRNGKey_map = random.split(PRNGKey, number_bins_over_lmin) # Prepare lmax+1-lmin PRNGKeys to be used

        # sampling_Wishart_map = jax.vmap(map_sampling_Wishart)(PRNGKey_map, jnp.arange(MICMAC_sampler_obj.lmin,MICMAC_sampler_obj.lmax+1))
        sampling_Wishart_map = jax.vmap(map_sampling_Wishart)(PRNGKey_map, jnp.arange(number_bins_over_lmin))
        # Map over PRNGKeys and ells to create samples of the Wishart distribution, of dimension [lmax+1-lmin,nstokes,nstokes]

        sampling_Wishart = sampling_Wishart.at[-number_bins_over_lmin:].set(sampling_Wishart_map)
        sampling_invWishart = jnp.linalg.pinv(sampling_Wishart)

        def reconstruct_spectra(ell):
            cond = jnp.logical_and(MICMAC_sampler_obj.bin_ell_distribution[:-1]<=ell, MICMAC_sampler_obj.bin_ell_distribution[1:]>ell)
            bin_contribution = jnp.where(cond, 1, 0)
            return jnp.einsum('lij,l->ij',sampling_invWishart,bin_contribution)/(ell*(ell+1))
        
        reconstructed_spectra = jax.vmap(reconstruct_spectra)(jnp.arange(MICMAC_sampler_obj.lmin,MICMAC_sampler_obj.lmax+1))
        return reconstructed_spectra


jitted_get_binned_inverse_wishart_sampling_from_c_ells_v2 = jax.jit(get_binned_inverse_wishart_sampling_from_c_ells_v2)
```
